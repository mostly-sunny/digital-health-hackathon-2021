```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/all-in-one.csv')
```

```python
dataset
```

```python
dataset.head()
```

```python
list_for_del=[]
```

```python
for i in range(1,11):
  for j in range(1000):
    if dataset['Var'+str(i)].values[j]>=10:
      list_for_del.append(j)
```

```python
for i in range(1000):
  if dataset['time'].values[i]<0:
    list_for_del.append(i)
```

```python
len(list_for_del)
```

```python
len(list(set(list_for_del)))
```

```python
list_for_del.sort()
```

```python
list_for_del
```

```python
dataset_modified = dataset.drop(list_for_del)
```

```python
dataset_modified
```

```python
dataset_modified.to_csv('/content/all-in-one-modified.csv', index= False)
```
